```python
import pandas as pd
import numpy as np
import math
 
first = pd.read_csv('data_house.csv').values 
length = len(first)
new = np.ones(length, dtype = float)
first =  first.astype('float')
first[:,0]=new 

train = first[range(1168),:]

test = first[range(1168,length),:]

print(train)
```

```python
minx1 = train[:,1].min()
maxx1 = train[:,1].max()
minx2 = train[:,2].min()
maxx2 = train[:,2].max()
minx3 = train[:,3].min()
maxx3 = train[:,3].max()
minx4 = train[:,4].min()
maxx4 = train[:,4].max()
minx5 = train[:,5].min()
maxx5 = train[:,5].max()
minx6 = train[:,6].min()
maxx6 = train[:,6].max()
minx7 = train[:,7].min()
maxx7 = train[:,7].max()

x1min = np.array([minx1 for x in range(1168)])
train[:,1] =(train[:,1] - x1min)/(maxx1-minx1)
x2min = np.array([minx2 for x in range(1168)])
train[:,2] =(train[:,2] - x2min)/(maxx2-minx2)
x3min = np.array([minx3 for x in range(1168)])
train[:,3] =(train[:,3] - x3min)/(maxx3-minx3)
x4min = np.array([minx4 for x in range(1168)])
train[:,4] =(train[:,4] - x4min)/(maxx4-minx4)
x5min = np.array([minx5 for x in range(1168)])
train[:,5] =(train[:,5] - x5min)/(maxx5-minx5)
x6min = np.array([minx6 for x in range(1168)])
train[:,6] =(train[:,6] - x6min)/(maxx6-minx6)
x7min = np.array([minx7 for x in range(1168)])
train[:,7] =(train[:,7] - x7min)/(maxx7-minx7)

x11min = np.array([minx1 for x in range(292)])
test[:,1] =(test[:,1] - x11min)/(maxx1-minx1)
x22min = np.array([minx2 for x in range(292)])
test[:,2] =(test[:,2] - x22min)/(maxx2-minx2)
x33min = np.array([minx3 for x in range(292)])
test[:,3] =(test[:,3] - x33min)/(maxx3-minx3)
x44min = np.array([minx4 for x in range(292)])
test[:,4] =(test[:,4] - x44min)/(maxx4-minx4)
x55min = np.array([minx5 for x in range(292)])
test[:,5] =(test[:,5] - x55min)/(maxx5-minx5)
x66min = np.array([minx6 for x in range(292)])
test[:,6] =(test[:,6] - x66min)/(maxx6-minx6)
x77min = np.array([minx7 for x in range(292)])
test[:,7] =(test[:,7] - x77min)/(maxx7-minx7)

xtrain = train[:,0:7]
ytrain = train[:,7].reshape(-1,1)

xtest = test[:,0:7]
ytest = test[:,7].reshape(-1,1)

print(xtest)
print(ytest)
```

```python
w = np.array([0. for x in range(7)]).reshape(-1,1)

loop = 1000000
cost = np.zeros((loop,1))
learning_rate = 0.5
for i in range(1,loop):
    r = np.dot(xtrain,w) - ytrain
    w[0] -= learning_rate*np.sum(r)/length
    cost[i] = 0.5*np.sum(r*r)/length
    w[1] -= learning_rate*np.sum(np.multiply(r, xtrain[:,1].reshape(-1,1)))/length
    w[2] -= learning_rate*np.sum(np.multiply(r, xtrain[:,2].reshape(-1,1)))/length
    w[3] -= learning_rate*np.sum(np.multiply(r, xtrain[:,3].reshape(-1,1)))/length
    w[4] -= learning_rate*np.sum(np.multiply(r, xtrain[:,4].reshape(-1,1)))/length
    w[5] -= learning_rate*np.sum(np.multiply(r, xtrain[:,5].reshape(-1,1)))/length
    w[6] -= learning_rate*np.sum(np.multiply(r, xtrain[:,6].reshape(-1,1)))/length
    if i%10000 == 0 : 
        print("gia tri cua cost:", cost[i],   "trong lan traing thu:", i)
    if i == 212000 :
        break
```

```python
denta = np.dot(xtest,w)-ytest
r = denta*denta
RMSE = math.sqrt(np.sum(r))
MAE = np.sum(abs(denta))/292
RAE = (np.sum(abs(denta)))/(np.sum(np.dot(xtest,w)))
print("gia tri RMSE:", RMSE)
print("gia tri MAE:", MAE)
print("gia tri RAE:", RAE)
```
